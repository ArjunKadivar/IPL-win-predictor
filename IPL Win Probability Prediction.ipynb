```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
match.shape
```

```python
delivery.head()
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df.head()
```

```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.head()
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
pipe2 = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe2.fit(X_train,y_train)
```

```python
lr_y_pred = pipe2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,lr_y_pred)
```

```python
params = {
    "step2__penalty" : ['l1', 'l2', 'elasticnet'],
    "step2__dual" : [True, False],
    "step2__tol" : [0.0001, 0.0002, 0.00005, 0.0003, 0.0004],
    "step2__C" : [1.0, 0.1, 1.01, 10],
    "step2__fit_intercept" : [True, False],
    "step2__solver" : ['newton-cg', 'lbfgs', 'liblinear', 'sag'],
    "step2__max_iter" : [100,200,300, 400],
    "step2__multi_class" : ['auto', 'ovr', 'multinomial']
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search = RandomizedSearchCV(pipe2, param_distributions=params, n_iter=5, scoring='roc_auc', n_jobs=-1, cv=5, verbose=3)
```

```python
random_search.fit(X_train,y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_params_
```

```python
clf = Pipeline(steps=[('step1',
                 ColumnTransformer(remainder='passthrough',
                                   transformers=[('trf',
                                                  OneHotEncoder(drop='first',
                                                                sparse=False),
                                                  ['batting_team',
                                                   'bowling_team', 'city'])])),
                ('step2',
                 LogisticRegression(C=0.1, fit_intercept=False, tol=5e-05))])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_lr = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_lr)
```

```python
clf.predict_proba(X_test)[2]
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(clf, X_train, y_train, cv=10)
```
